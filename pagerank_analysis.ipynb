```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Hyperlink Matrix
Hl_matrix = np.array([np.array([0, 1/2, 1/2, 0, 0 , 0]),
             np.array([0, 0, 1, 0, 0, 0]),
             np.array([1/3, 1/3, 0, 0, 1/3, 0]),
             np.array([0, 0, 0 , 0, 1/2, 1/2]),
             np.array([0, 0, 0, 1/2, 0, 1/2]),
             np.array([0, 0, 0, 1, 0, 0])])
Hl_matrix = Hl_matrix.T
print(Hl_matrix)

rank_vector = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6]).T
```

```python
# Iterations of vector x after dot product
Hl_matrix = np.asarray(Hl_matrix)
rank_vector = np.asarray(rank_vector)

list_of_x = []
for i in np.linspace(1, 50):
    x_new = np.dot(Hl_matrix, rank_vector)
    rank_vector = x_new
    list_of_x.append(x_new)

print(x_new)
```

```python
X = np.array(list_of_x)  # shape: (num_iterations, vector_length)
print(X.shape)

for i in range(X.shape[1]):  # iterate over vector components
    plt.plot(X[:, i], label=f'Component {i+1}')

plt.xlabel('Iteration')
plt.ylabel('Value')
plt.title('Change in Vector Components Over Iterations')
plt.legend()
plt.show()
```

```python
H = np.array([np.array([0, 1/2, 1/2, 0, 0 , 0]),
             np.array([0, 0, 1, 0, 0, 0]),
             np.array([1/3, 1/3, 0, 0, 1/3, 0]),
             np.array([0, 0, 0 , 0, 1/2, 1/2]),
             np.array([0, 0, 0, 1/2, 0, 1/2]),
             np.array([0, 0, 0, 1, 0, 0])])

H = H.T

# Compute eigenvalues and eigenvectors
eigvals, eigvecs = np.linalg.eig(H)

# Find index of the largest eigenvalue
max_index = np.argmax(eigvals)

# Get the corresponding eigenvector
principal_eigvec = eigvecs[:, max_index]

# (Optional) Normalize the eigenvector
principal_eigvec = principal_eigvec / np.sum(principal_eigvec)

print("Eigenvalues:", eigvals)
print("Principal Eigenvalue:", eigvals[max_index])
print("Principal Eigenvector (normalized):", principal_eigvec.real)
```
